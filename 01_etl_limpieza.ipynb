```python
import pandas as pd
import numpy as np
import os
from datetime import datetime

# Configurar pandas para mostrar más columnas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# Rutas (ajusta la carpeta raw donde tengas los CSVs)
raw_path = 'data/raw/'
processed_path = 'data/processed/'
os.makedirs(processed_path, exist_ok=True)

# Lista de archivos a cargar
files = {
    'customers': 'olist_customers_dataset.csv',
    'geolocation': 'olist_geolocation_dataset.csv',
    'order_items': 'olist_order_items_dataset.csv',
    'order_payments': 'olist_order_payments_dataset.csv',
    'order_reviews': 'olist_order_reviews_dataset.csv',
    'orders': 'olist_orders_dataset.csv',
    'products': 'olist_products_dataset.csv',
    'sellers': 'olist_sellers_dataset.csv',
    'category_translation': 'product_category_name_translation.csv'  # opcional
}

# Cargar cada archivo en un diccionario de DataFrames
dfs = {}
for name, file in files.items():
    path = os.path.join(raw_path, file)
    if os.path.exists(path):
        dfs[name] = pd.read_csv(path)
        print(f"Cargado: {name} -> {dfs[name].shape}")
    else:
        print(f"No encontrado: {path}")
```

```python
# Ver información general de cada tabla
for name, df in dfs.items():
    print(f"\n=== {name} ===")
    print(f"Shape: {df.shape}")
    print(f"Columnas: {list(df.columns)}")
    print(f"Nulos por columna:\n{df.isnull().sum()}")
    print(f"Duplicados fila completas: {df.duplicated().sum()}")
    # Mostrar primeras filas
    display(df.head(2))
```

```python
# Trabajar sobre copia para no modificar original
orders = dfs['orders'].copy()

# Columnas de fecha
date_cols = ['order_purchase_timestamp', 'order_approved_at', 
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    if col in orders.columns:
        orders[col] = pd.to_datetime(orders[col], errors='coerce')  # errores->NaT

# Revisar estado de pedidos
print(orders['order_status'].value_counts())

# Guardar limpias (sobrescribimos en dfs)
dfs['orders'] = orders

display(orders.head(2))
```

```python
items = dfs['order_items'].copy()

# Convertir precio y flete a float
items['price'] = pd.to_numeric(items['price'], errors='coerce')
items['freight_value'] = pd.to_numeric(items['freight_value'], errors='coerce')

# Ver si hay precios negativos o cero
print(f"Precios <= 0: {(items['price'] <= 0).sum()}")
# Opcional: eliminar filas con precio <= 0 (son datos erróneos)
items = items[items['price'] > 0]

dfs['order_items'] = items

display(items.head(2))
```

```python
payments = dfs['order_payments'].copy()
payments['payment_value'] = pd.to_numeric(payments['payment_value'], errors='coerce')
payments['payment_installments'] = pd.to_numeric(payments['payment_installments'], errors='coerce')

# Agrupar pagos por order_id (un pedido puede tener múltiples formas de pago)
# Esto lo haremos más adelante. Por ahora solo limpiar nulos.
# Eliminar filas sin payment_value
payments = payments.dropna(subset=['payment_value'])

dfs['order_payments'] = payments
display(payments.head(2))
```

```python
reviews = dfs['order_reviews'].copy()
reviews['review_score'] = pd.to_numeric(reviews['review_score'], errors='coerce')

# Convertir fecha de respuesta si existe
if 'review_answer_timestamp' in reviews.columns:
    reviews['review_answer_timestamp'] = pd.to_datetime(reviews['review_answer_timestamp'], errors='coerce')

dfs['order_reviews'] = reviews
display(reviews.head(2))
```

```python
customers = dfs['customers'].copy()
sellers = dfs['sellers'].copy()

# Ver nulos (no debería haber)
print("Clientes nulos:\n", customers.isnull().sum())
print("Vendedores nulos:\n", sellers.isnull().sum())

# Ambos tienen zip_code_prefix como string, lo dejamos así.

dfs['customers'] = customers
dfs['sellers'] = sellers
display(customers.head(2))
display(sellers.head(2))
```

```python
products = dfs['products'].copy()

# Las columnas numéricas (altura, peso, etc.) pueden tener nulos. Los dejamos por ahora.
numeric_prod = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
for col in numeric_prod:
    products[col] = pd.to_numeric(products[col], errors='coerce')

# Si hay categoría nula, la etiquetamos como 'desconocida'
products['product_category_name'] = products['product_category_name'].fillna('unknown')

dfs['products'] = products
display(products.head(2))
```

```python
geo = dfs['geolocation'].copy()

# Quedarnos con la primera ocurrencia de cada zip_code_prefix (la más común)
geo = geo.drop_duplicates(subset=['geolocation_zip_code_prefix'], keep='first')

dfs['geolocation'] = geo
display(geo.head(2))
```

```python
if 'category_translation' in dfs:
    cat_trans = dfs['category_translation'].copy()
    # Limpiar nombres
    cat_trans['product_category_name_english'] = cat_trans['product_category_name_english'].str.lower().str.strip()
    dfs['category_translation'] = cat_trans
    display(cat_trans.head(2))
```

## Validación de valores nulos críticos


```python
# Verificar que las claves foráneas no tengan nulos donde no deberían
critical_checks = [
    (orders, 'order_id', 'No debe haber nulos en order_id'),
    (orders, 'customer_id', 'No debe haber nulos en customer_id'),
    (items, 'order_id', 'No debe haber nulos en order_id'),
    (items, 'product_id', 'No debe haber nulos en product_id'),
    (items, 'seller_id', 'No debe haber nulos en seller_id')
]

for df, col, msg in critical_checks:
    nulos = df[col].isnull().sum()
    if nulos > 0:
        print(f"⚠️ {msg}: {nulos} nulos")
    else:
        print(f"✅ {msg}")
```

```python
# Guardar cada tabla limpia en processed
for name, df in dfs.items():
    output_file = os.path.join(processed_path, f'{name}_clean.csv')
    df.to_csv(output_file, index=False)
    print(f"Guardado: {output_file} ({df.shape})")
```
