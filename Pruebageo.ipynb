```python
import pandas as pd

raw_path = 'data/raw/'

geo = pd.read_csv(raw_path + "olist_geolocation_dataset.csv")

geo.head()
```

```python
geo.columns
```

```python
geo[['geolocation_lat', 'geolocation_lng']].describe()
```

```python
import plotly.express as px

fig = px.scatter_geo(
    geo.sample(10000),  # para no saturar
    lat='geolocation_lat',
    lon='geolocation_lng',
    hover_name='geolocation_city',
    color='geolocation_state',
    title='Geolocalización de pedidos en Brasil'
)

fig.update_layout(
    geo_scope='south america'
)

fig.show()
```

```python
fig = px.density_mapbox(
    geo.sample(10000),
    lat='geolocation_lat',
    lon='geolocation_lng',
    radius=8,
    zoom=3,
    mapbox_style='carto-positron'
)

fig.show()
```
